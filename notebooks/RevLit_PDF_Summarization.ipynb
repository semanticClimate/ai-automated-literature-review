## **Step 5:Loading the PDF**

## *User input (PDF file path ) required to load and summarize.*

```python
import os

# 🔄 USER INPUT: Either a single PDF file or a folder path
input_path = "/content/downloaded_file"  # Replace with your own file or folder path

# Collect PDF paths based on whether input is a file or a folder
pdf_paths = []
if os.path.isfile(input_path) and input_path.endswith(".pdf"):
    pdf_paths = [input_path]  # Single PDF
elif os.path.isdir(input_path):
    for root, dirs, files in os.walk(input_path):
        for file in files:
            if file.endswith(".pdf"):
                pdf_paths.append(os.path.join(root, file))
else:
    print("❌ Invalid path. Please provide a PDF file or folder containing PDFs.")

print(f"✅ Found {len(pdf_paths)} PDF(s) to process.")
```

### **Step 6: Import Libraries,Initialize the Tokenizer & Model**
While multiple models are available for this task, we primarily use *sshleifer/distilbart-cnn-12-6*, alongside you can try other models such as:

* **facebook/bart-large-cnn**

* **nsi319/legal-pegasus**

* **google/pegasus-pubmed**
* You can explore even more models and benchmark them for your specific use case via the[Hugging Face Summarization Models](https://huggingface.co/models?pipeline_tag=summarization&sort=trending)


```python
import fitz # PyMuPDF for extracting text from PDFs
import nltk
import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Download necessary NLTK resources
nltk.download('punkt')

# Load model and tokenizer from Huggingface
checkpoint = "sshleifer/distilbart-cnn-12-6" #facebook/bart-large-cnn #nsi319/legal-pegasus # google/pegasus-pubmed
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
max_input_length = tokenizer.model_max_length

def remove_references_section(text):
    # Keywords that may indicate the start of references section
    ref_keywords = ["references", "bibliography", "works cited"]
    lines = text.split("\n")
    cleaned_lines = []
    for line in lines:
        if any(keyword in line.strip().lower() for keyword in ref_keywords):
            break  # stop reading text after hitting references section
        cleaned_lines.append(line)
    return "\n".join(cleaned_lines)

# Function to extract DOI using regex
def extract_doi(text):
    doi_pattern = r'\b10\.\d{4,9}/[-._;()/:A-Z0-9]+'
    matches = re.findall(doi_pattern, text, flags=re.IGNORECASE)
    return matches[0] if matches else "DOI not found"

def summarize_pdf(pdf_path, max_output_length=50):
    # Extract text from PDF
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()

     # Try extracting DOI before removing anything
    doi = extract_doi(text)
    # Remove references
    cleaned_text = remove_references_section(text)
    # Tokenize into sentences
    sentences = nltk.tokenize.sent_tokenize(text)

    # Chunk into manageable size
    chunks = []
    length = 0
    chunk = ""
    for sentence in sentences:
        sentence_length = len(tokenizer.tokenize(sentence))
        if length + sentence_length <= max_input_length:
            chunk += sentence + " "
            length += sentence_length
        else:
            chunks.append(chunk.strip())
            chunk = sentence + " "
            length = sentence_length
    if chunk:
        chunks.append(chunk.strip())

    # Summarize each chunk
    summaries = []
    for chunk in chunks:
        inputs = tokenizer(chunk, return_tensors="pt", truncation=True, padding=True, max_length=max_input_length)
        output = model.generate(
            **inputs,
            max_length=max_output_length,
            min_length=10,
            length_penalty=0.8,
            early_stopping=True,
            no_repeat_ngram_size=3
        )
        decoded = tokenizer.decode(output[0], skip_special_tokens=True)

        # Clean summary
        cleaned = re.sub(r'\b(?:figure|fig)\s*\d+', '', decoded)
        cleaned = re.sub(r'\[\d+(?:,\d+)*\]', '', cleaned)
        cleaned = re.sub(r'\d+(\.\d+)+', '', cleaned)

        summaries.append(cleaned.strip())

    return doi," ".join(summaries)
```

### **Step 7: Generate Summaries**

```python
import nltk
nltk.download('punkt_tab')
summaries_dict = {}

for i, pdf_path in enumerate(pdf_paths, 1):
    print(f"Summarizing paper {i}/{len(pdf_paths)}: {os.path.basename(pdf_path)}")
    try:
        doi, summary = summarize_pdf(pdf_path)  # Get both DOI and summary
        summaries_dict[pdf_path] = {"DOI": doi, "Summary": summary}
        print(f" DOI: {doi}\n Summary: {summary[:200]}...")
        #summaries_dict[pdf_path] = summary
    except Exception as e:
        print(f"Error summarizing {pdf_path}: {e}")
```

### **Step 8:Display the Final Summary & Export Summarized Results to CSV.** (For direct Pipeline[CLICK HERE](https://colab.research.google.com/drive/159zAppwWtrKaMP5yVg8Cwm7wfOz0MIKm?usp=sharing))

```python
import csv
import os
import pandas as pd
from IPython.display import display, HTML

def clean_text(text):
    if not isinstance(text, str):
        return text
    return text.replace("\\n", " ").replace("\n", " ").strip()

with open("summarized_papers.csv", "w", newline='', encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Folder Name", "DOI Link", "Summary"])

    for pdf_file, content in summaries_dict.items():
        folder_name = os.path.basename(os.path.dirname(pdf_file))
        doi = clean_text(content["DOI"])

        doi_link = f"https://doi.org/{doi}" if doi != "DOI not found" else doi
        summary = clean_text(content["Summary"])

        writer.writerow([folder_name, doi_link, summary])

# Load your CSV
df = pd.read_csv("summarized_papers.csv")

# Clean up any literal '\n' or newline characters
for col in ['Folder Name', 'DOI Link', 'Summary']:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace(r'\\n', ' ', regex=True)    # Handles '\\n'
    df[col] = df[col].str.replace(r'\n', ' ', regex=True)     # Handles '\n'
    df[col] = df[col].str.replace(r'\\\\n', ' ', regex=True)  # Handles '\\\\n'
    df[col] = df[col].str.strip()

# Pre-process the summary for safe insertion (replace newlines with <br> before the f-string)
df['SummaryHTML'] = df['Summary'].apply(lambda x: x.replace('\n', '<br>'))

# Combine content into one cell with proper formatting and line breaks
df['Combined'] = df.apply(lambda row: (
    "<div style='font-family: Arial, sans-serif; font-size: 15px; line-height: 1.5; text-align: left;'>"
    f"<strong>PMCID:</strong> {row['Folder Name']}<br>"
    f"<strong>DOI:</strong> <a href='{row['DOI Link']}' target='_blank'>{row['DOI Link']}</a><br>"
    "<details>"
    "<summary style='margin-top: 4px;'>Click to view summary</summary>"
    f"<div style='margin-top: 5px;'>{row['SummaryHTML']}</div>"
    "</details></div>"
), axis=1)

# HTML output with styling
html_output = df[['Combined']].to_html(escape=False, index=False, header=False)

# Add styling for left alignment
custom_style = """
<style>
table {
    border-collapse: collapse;
    width: 80%;
    font-family: Arial, sans-serif;
    text-align: left;
}
td {
    border: 1px solid #ccc;
    padding: 10px;
    vertical-align: top;
    text-align: left;
}
</style>
"""

# Display the final HTML
display(HTML(custom_style + html_output))
```
